```python
import pandas as pd
import os
```

```python
path = "/content/drive/MyDrive/Dasprog"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files
all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file, delimiter=';', usecols=["ID","Product","Price","QTY","Date","Address"])
    all_months_data = pd.concat([all_months_data, current_data])

all_months_data.to_csv("all_data_copy.csv", index=False)
```

```python
all_data = pd.read_csv("all_data_copy.csv")
# Find NAN
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())

all_data = all_data.dropna(how='all')
all_data
```

```python
all_data['Month'] = all_data['Date'].str[3:5]
all_data['Month'] = all_data['Month'].astype('int32')
all_data
```

```python
all_data['Sales'] = all_data["QTY"]*all_data["Price"]
all_data
```

```python
all_data.groupby('Month').sum()
```

```python
product_group = all_data.groupby('Product')
quantity_order = product_group.sum()["QTY"]
quantity_order.head()
```

```python
sales_product = all_data['Sales'].groupby(all_data['Product']).sum()
sales_product.head()
```

```python
mean_sales = all_data['Sales'].mean()
mean_sales
```
